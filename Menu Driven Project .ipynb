```python
import requests
import bs4
from geopy.geocoders import Nominatim
from gtts import gTTS
import os
from ipaddress import IPv4Address
from pyairmore.request import AirmoreSession
from pyairmore.services.messaging import MessagingService
from twilio.rest import Client
import smtplib
def show_menu():
    print("\n-----------Menu Driven Project-------------")
    print("\n\t--------MENU-------")
    print("1. Scrape data from Google")
    print("2. Find geographical locations and coordinates")
    print("3. Convert text to voice")
    print("4. Send SMS to a mobile number")
    print("5. Connect to a mobile phone and send a message")
    print("6. Send an email")
    print("7. Send batch emails")
    print("0. Exit")
def handle_choice(choice):
    if choice == "1":
        scraping_data()
    elif choice == "2":
        find_geographical_location()
    elif choice == "3":
        convert_text_to_voice()
    elif choice == "4":
        send_sms()
    elif choice == "5":
        send_sms_to_android()
    elif choice == "6":
        send_email()
    elif choice == "7":
        send_batch_emails()
    elif choice == "0":
        print("Exiting the program. Goodbye!")
        return False
    else:
        print("Invalid choice. Please enter a valid option.")
        return True
def main():
    while True:
        show_menu()
        choice = input("Enter a choice from the given menu: ")
        handle_choice(choice)
def scraping_data():
    txt = input("Enter the text to search in Google: ")
    url = 'https://google.com/search?q=' + txt
    try:
        res = requests.get(url)
        res.raise_for_status()  
        s = bs4.BeautifulSoup(res.text, "html.parser")
        headings = s.find_all('h3')
        if headings:
            print("Search Results:")
            for info in headings:
                print(info.getText())
        else:
            print("No results found.")
    except requests.exceptions.RequestException as e:
        print(f"Error occurred while scraping data: {e}")
def find_geographical_location():
    location = Nominatim(user_agent="GetLoc")
    district = input("Enter your District: ")
    locality = input("Enter your Locality: ")
    try:
        getLoc = location.geocode(f"{locality}, {district}")
        if getLoc:
            print("Address:", getLoc.address)
            print("Latitude:", getLoc.latitude)
            print("Longitude:", getLoc.longitude)
        else:
            print("Location not found. Please check your input.")
    except Exception as e:
        print(f"Error occurred while finding location: {e}")
def convert_text_to_voice():
    txt = input("Enter the text: ")
    language = "en"
    try:
        obj = gTTS(text=txt, lang=language, slow=False)
        obj.save("voice.mp3")
        os.system("start voice.mp3")
        print("Voice conversion successful. Playing audio...")
    except Exception as e:
        print(f"Error occurred during text-to-voice conversion: {e}")
def send_sms_to_android():
    ip = input("Enter the IP address of your Android device: ")
    port = 2333
    try:
        session = AirmoreSession(IPv4Address(ip), port)
        acc = session.request_authorization()
        if acc:
            service = MessagingService(session)
            phone_number = input("Enter the recipient's phone number: ")
            message = input("Enter the message to send: ")
            service.send_message(phone_number, message)
            print("Message sent successfully!")
        else:
            print("Authorization failed. Please check your connection.")
    except Exception as e:
        print(f"Error occurred while sending SMS to Android: {e}")
def send_sms():
    sid = "AC113719d91bdee7deb1f09256c967c4d3"
    auth = "5265cd23ec821e85e7ff19adc7127e9c"
    tnum = '+13344296405'
    rnum = input("Enter the recipient's phone number: ")
    client = Client(sid, auth)
    message_body = input("Enter the message to send: ")
    try:
        message = client.messages.create(body=message_body, from_=tnum, to=rnum)
        print("Message sent successfully!")
    except Exception as e:
        print(f"Error occurred while sending SMS: {e}")
def send_email():
    sender = input("enter the senders Email:")
    reciever = input("enter the reciever Email:")
    subject = input("Enter the subject:")
    message = input("enter the Message:")
    txt = f"Subject:{subject}\n\n{message}"
    server = smtplib.SMTP("smtp.gmail.com",587)
    server.starttls()
    server.login(sender ,"Enter Your Google app password")
    server.sendmail(sender , reciecer , txt)
    print("receiver",reciever)
if __name__ == "__main__":
    main()
```
